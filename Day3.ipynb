```python
class student:
  def __init__(self, name, num):
    self.name = name
    self.num = num
  def display(self):
    print(f"Name:{self.name}, RollNo:{self.num}")

s1 = student('Tony', 101)
s2 = student('Mike', 103)
s1.display()
s2.display()
```

```python
#code for calculating the area and perimeter of a rectangle by taking a user defined input as an object and class as rectangle, return the values to display
class rectangle:
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth
  def area(self):
    return self.length * self.breadth
  def peri(self):
    return 2*self.length + 2*self.breadth

length = int(input("enter the value of length:"))
breadth = int(input("enter the value of breadth:"))
r = rectangle(length, breadth)
print("Area of rectangle is:", r.area())
print("Perimeter of rectangle is:", r.peri())
```

```python
# calling a method from another method in a single class

class abc():
  def __init__(self, var):
    self.var=var
  def display(self):
    print("var:", self.var)
  def call(self):
    self.var+=1000
    self.display()
obj = abc(100)
obj.call()
```

```python
# code for a class method to call a function, which is defined in global namesplace
def globe(x):
  return x*10
class abc():
  def __init__(self, var):
    self.var = var
  def display(self):
    print("Var:", self.var)
  def modify(self):
    self.var= globe(self.var)

obj = abc(10)
obj.display()
obj.modify()
obj.display()
```

```python
'''program that uses class as student to store the name and marks
of the student, use a list to store the marks of 3 subjects
contraints:
1.take class as student
2.create a constructor for the student name
3.create a function for marks, to be entered manually with in that class function and add the marks to the list
4.display then student name and the marks he/she got
5.pass the object's attributes of two student names
tc:
obj1:"vijay"
obj2:"Anil"

output: vijay got[88,88,90]
        Anil got[77,78,90]
'''




class student:
  def name(self,name):
    self.name=name
  def marks(self,marks):
    self.marks = marks
  def display(self):
    print(self.name,"got",self.marks)

obj1 = student()
obj2 = student()
obj1.name("vijay")
obj1.marks([88,88,90])
obj2.name("Anil")
obj2.marks([77,78,90])
obj1.display()
obj2.display()
```

```python
class student:
  def __init__(self,name):
    self.name=name
    self.marks=[]
  def entermarks(self):
    for i in range(3):
      m = int(input("Enter the marks of %s in object %d: "%(self.name, i+1)))
      self.marks.append(m)
  def display(self):
    print(self.name, "got", self.marks)

s1 = student("vijay")
s1.entermarks()
s2 = student("Anil")
s2.entermarks()
s1.display()
s2.display()
```
